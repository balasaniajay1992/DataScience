# **Numpy Slicing and Indexing**

```python
x=np.arange(24).reshape((6,4))
x
```

```python
x[3] # using index we can access one row at a time
```

```python
x[3][3] # row with element
```

```python
print(x[3,3])
```

```python
print(x[4,2],x[5,2])
```

```python
# numpy array is mutable in nature
x[3,3]
```

```python
# replace 15 by 50
x[3,3]=50
x
```

```python
# access more rows
x[:] #[start_index:stop_index]
```

```python
x[1:4]
```

```python
x[3:] # or x[-3:] +ve indexing and -ve indexing will work in indexing
```

```python
x[::-2]
```

```python
# column operation
x[-2:,-2:] # [row,column]
```
